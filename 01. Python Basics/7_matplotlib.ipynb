```python
#Kaggle sitesinden bulduğunuz veri kümesini Pandas'ı kullanarak bir veri çerçevesine aktarın. 
#Veri kümenizde yer alan kategorik değişkenlerden iki tanesini seçin ve subplot kullanarak, değişkenlerin çubuk grafiklerini çıkarın.

%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('train.csv',index_col="PassengerId")
df = df.dropna(subset=['Age'])

plt.figure(figsize=(20,10))

#### 1. Grafik ####
plt.subplot(2, 2, (1,2))

dfd = pd.DataFrame()
dfd['Yolcu Sayısı'] = list(df.groupby(by='Age').count()['Name'])
dfd['Yolcu Yaşı'] = list(dict.fromkeys(df.Age[df.Age.notna()].sort_values()))

y = dfd['Yolcu Sayısı']
x_labels = dfd['Yolcu Yaşı']
x_pos = np.arange(len(y))

plt.bar(x_pos,y,alpha=.9)
#plt.xticks(x_pos,x_labels,rotation='vertical',fontsize=6)

plt.title("Titanic - Yolcu / Yaş Dağılımı")
plt.xlabel("Yolcu Yaşı")
plt.ylabel("Yolcu Sayısı")

#### 2. Grafik ####
plt.subplot(2, 2, 3)
df1=df[df['Survived']==1]

dfd = pd.DataFrame()
dfd['Yolcu Sayısı'] = list(df1.groupby(by='Age').count()['Name'])
dfd['Yolcu Yaşı'] = list(dict.fromkeys(df1.Age[df1.Age.notna()].sort_values()))

y = dfd['Yolcu Sayısı']
x_labels = dfd['Yolcu Yaşı']
x_pos = np.arange(len(y))

plt.bar(x_pos,y,color="red",alpha=.5)
#plt.xticks(x_pos,x_labels,rotation='vertical',fontsize=7)

plt.title("Titanic - Hayatta Kalan Yolcu / Yaş Dağılımı")
plt.xlabel("Yolcu Yaşı")
plt.ylabel("Yolcu Sayısı")

#### 3. Grafik ####
plt.subplot(2, 2, 4)
df0=df[df['Survived']==0]

dfd = pd.DataFrame()
dfd['Yolcu Sayısı'] = list(df0.groupby(by='Age').count()['Name'])
dfd['Yolcu Yaşı'] = list(dict.fromkeys(df0.Age[df0.Age.notna()].sort_values()))

y = dfd['Yolcu Sayısı']
x_labels = dfd['Yolcu Yaşı']
x_pos = np.arange(len(y))

plt.bar(x_pos,y,color="purple",alpha=.5)
#plt.xticks(x_pos,x_labels,rotation='vertical',fontsize=7)

plt.title("Titanic - Hayatını Kaybeden Yolcu / Yaş Dağılımı")
plt.xlabel("Yolcu Yaşı")
plt.ylabel("Yolcu Sayısı")

plt.tight_layout()
plt.show()
```
