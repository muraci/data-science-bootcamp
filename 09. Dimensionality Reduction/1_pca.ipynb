_Fashion-MNIST veri kümesini kullanarak:_
- _Veri setini yükleyin ve verileri normalleştirmek gibi ön işlemlerinizi (preprocessing) yapın. Verilere PCA uygulayın ve ilk iki ana bileşeni elde edin._
- _İlk iki ana bileşeni kullanarak, derste yaptığımız gibi veri noktalarını renklendirerek ve sınıflandırarak veri grafiğini çizin._
- _PCA çözümünün tatmin edici olduğunu düşünüyor musunuz? Farklı sınıfları kolayca ayırt edebiliyor musunuz?_

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
mnist = fetch_openml('Fashion-MNIST', version = 1, cache = True)
```

```python
np.random.seed(42)

idx = np.random.choice(70000,10000)
x = mnist.data[idx]
y = mnist.target[idx].astype(np.int64)

x_std = StandardScaler().fit_transform(x)
```

```python
x_pca = PCA(n_components = 2).fit_transform(x_std)

plt.figure(figsize=(16,8))
plt.scatter(x_pca[:, 0], x_pca[:, 1], alpha = 0.7, c = y)
plt.axis('off')
plt.show()
```

```python
x_pca = PCA(n_components = 3).fit_transform(x_std)

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_pca[:, 0], x_pca[:, 1], x_pca[:, 2], c = y)
plt.show()
```
